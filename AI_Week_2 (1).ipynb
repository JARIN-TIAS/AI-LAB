#BFS

```python
graph = {
 '5' : ['3','7'],
 '3' : ['2', '4'],
 '7' : ['8'],
 '2' : [],
 '4' : ['8'],
 '8' : []
}
visited = [] # List for visited nodes.
queue = [] #Initialize a queue
```

```python
def bfs(visited, graph, node): #function for BFS
    visited.append(node)
    queue.append(node)
    while queue: # Creating loop to visit each node
        m = queue.pop(0)
        print (m, end = " ")
        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
```

```python
# Driver Code
print("Following is the Breadth-First Search")
bfs(visited, graph, '5') # function calling
```

If you start BFS from node '5':

Initially, visited = [] and queue = [].
You add '5' to both visited and queue. Now, visited = ['5'] and queue = ['5'].
The first iteration of the while loop processes '5', printing it.
'5''s neighbors are '3' and '7'. Both are added to visited and queue, so now visited = ['5', '3', '7'] and queue = ['3', '7'].
Next, '3' is dequeued, and its neighbors '2' and '4' are added to visited and queue. Now, visited = ['5', '3', '7', '2', '4'] and queue = ['7', '2', '4'].
Then, '7' is processed, and its neighbor '8' is added to visited and queue. Now, visited = ['5', '3', '7', '2', '4', '8'] and queue = ['2', '4', '8'].
'2', '4', and '8' are processed in sequence, but they have no new neighbors, so the loop ends.
Conclusion:
At the end of the BFS traversal, all nodes reachable from the starting node '5' are printed in the order they were visited. The BFS function ensures that each node is visited once and that neighbors are explored level by level

```python
class Solution:
    def solve(self, board):
        state_map = {}  # Dictionary to track visited states and their depths
        flatten = tuple([num for row in board for num in row])  # Flatten the board

        # Add the initial state to the map
        state_map[flatten] = 0

        # If the initial state is already the goal
        if flatten == (0, 1, 2, 3, 4, 5, 6, 7, 8):
            return 0

        # Start BFS to find the solution
        return self.bfs(state_map)

    def bfs(self, state_map):
        depth = 0

        while True:
            # Get the current states at this depth
            current_states = [state for state in state_map if state_map[state] == depth]

            # If no more states to explore, return -1 (unsolvable)
            if not current_states:
                return -1

            # Explore each state at the current depth
            for state in current_states:
                next_states = self.find_next(state)  # Generate next possible moves
                for next_state in next_states:
                    # If the state is new, add it to the map with incremented depth
                    if next_state not in state_map:
                        state_map[next_state] = depth + 1

                        # If the goal state is reached, return the depth
                        if next_state == (0, 1, 2, 3, 4, 5, 6, 7, 8):
                            return depth + 1

            # Increment depth for the next level
            depth += 1

    def find_next(self, state):
        # Define possible moves for each position of the blank tile (0)
        moves = {
            0: [1, 3],
            1: [0, 2, 4],
            2: [1, 5],
            3: [0, 4, 6],
            4: [1, 3, 5, 7],
            5: [2, 4, 8],
            6: [3, 7],
            7: [4, 6, 8],
            8: [5, 7],
        }

        # Find the current position of the blank tile
        pos_0 = state.index(0)
        results = []

        # Generate new states by swapping the blank tile with its neighbors
        for move in moves[pos_0]:
            new_state = list(state)
            new_state[move], new_state[pos_0] = new_state[pos_0], new_state[move]
            results.append(tuple(new_state))

        return results


# Example usage
ob = Solution()
matrix = [
    [3, 1, 2],
    [4, 7, 5],
    [6, 8, 0]
]
print(ob.solve(matrix))  # Output: Minimum moves to solve the puzzle
```
